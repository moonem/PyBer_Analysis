```python
# import dependencies.
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the x, y limit between 0 and 45.
plt.xlim(-1, 12)
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by month")
# Add gridline.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create Labels for x, y axes.
plt.xlabel("Date  ")
plt.ylabel("Fare($)")
# Create a Title.
plt.title("PyBer Fare by Month")
# Add the Legend.
plt.legend()
```

```python
# Create a horizontal plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")

plt.legend()
```

```python
# Object oriented approach.
# Create the plot with ax.plot().
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
# Use scatter plot with plt.plot.
plt.plot(x_axis, y_axis, "o")
```

```python
# Use scatter plot.
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.ylabel("Date  ")
plt.xlabel("Fare($)")
# Create a Title.
plt.title("PyBer Fare by Month")
# Add the Legend.
plt.legend()
# Invert y-axis.
plt.gca().invert_yaxis()
plt.show()
```

```python
# Use scatter plot to create Bubble chart.
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis], c='green', label='Chicago')
plt.xlabel("Date  ")
plt.ylabel("Fare($)")
# Create a Title.
plt.title("PyBer Fare by Month")
# Add the Legend.
plt.legend()
# Invert y-axis.
#plt.gca().invert_yaxis()
plt.show()
```

```python
# Scatter plot using object-oriented interface.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='skyblue', alpha=0.8, linewidth=2, edgecolor='k', label='Boston')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50) # set x-axis limit (0, 50)
ax.invert_yaxis() # invert y-axis value (months from Jan-Dec)
```

```python
# Pie chart.
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0) # we've to match for all the x-values
plt.subplots(figsize=(8, 8)) # define the size of the pie
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%') # %formatting

plt.show()
```

```python
# Pie chart using Object-oriented interface.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0) # we've to match for all the x-values
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%',
        shadow=True,
        startangle=-270,
        counterclock=False)        
plt.show()
```
